```python
import pandas as pd
```

```python
TopBeers = pd.DataFrame(pd.read_csv("TopBeers.csv"))
BreweryGeo = pd.DataFrame(pd.read_csv("breweries_geocode.csv"))
BreweryID = pd.DataFrame(pd.read_csv("breweries.csv"))
```

```python
TopBeers.head()
```

```python
BreweryID.head()
```

```python
BreweryGeo.head()
```

```python
GeoCodes = pd.merge(BreweryGeo,BreweryID, how='inner', on='id')
```

```python
GeoCodes
```

```python
TopBeers = TopBeers.rename(columns={'name':'beername','brewery':'name'})
```

```python
GeoCodesAll = pd.merge(GeoCodes, TopBeers, how='inner', on='name')
```

```python
GeoCodesAll
```

```python
GeoCodesAll.to_csv("BeerMapPoints.csv")
```
